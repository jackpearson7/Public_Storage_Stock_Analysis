Question 1: How has Public Storage’s stock and the stock of their competitors performed each day over the time period in the data set?

Business Justification: This query will give an overall look at Public Storage's stock performance in comparison to its competitors. This data can be used for surface level buisness planning.

SQL Features: CASE, GROUP BY, JOIN

```python
%%sql
SELECT LEFT(psa_data.`Date`, 10) AS Date, 
	CASE
		WHEN SUM(psa_data.`Close`-psa_data.`Open`) = 0 THEN 'No Change'
		WHEN SUM(psa_data.`Close`-psa_data.`Open`) > 0 THEN 'Up'
		ELSE 'Down'
	END AS PSA_Daily_Movement,
	CONCAT(ROUND(SUM(((psa_data.`Close`-psa_data.`Open`)/psa_data.`Open`*100)), 2), '%') AS PSA_Daily_Change, 
	CASE
		WHEN SUM(cube_data.`Close`-cube_data.`Open`) = 0 THEN 'No Change'
		WHEN SUM(cube_data.`Close`- cube_data.`Open`) > 0 THEN 'Up'
		ELSE 'Down'
	END AS CUBE_Daily_Movement,
	CONCAT(ROUND(SUM(((cube_data.`Close`-cube_data.`Open`)/cube_data.`Open`*100)), 2), '%') AS CUBE_Daily_Change, 
	CASE
		WHEN SUM(lsi_data.`Close`-lsi_data.`Open`) = 0 THEN 'No Change'
		WHEN SUM(lsi_data.`Close`- lsi_data.`Open`) > 0 THEN 'Up'
		ELSE 'Down'
	END AS LSI_Daily_Movement,
	CONCAT(ROUND(SUM(((lsi_data.`Close`-lsi_data.`Open`)/lsi_data.`Open`*100)), 2), '%') AS LSI_Daily_Change
FROM psa_data
JOIN cube_data
	ON psa_data.`Date` = cube_data.`Date`
JOIN lsi_data
	ON psa_data.`Date` = lsi_data.`Date`
GROUP BY psa_data.`Date`;
```

Question 1 Recommendations: The results from this query can be used to get a quick snapshot of Public Storage's performance over time in relation to its competitors. If the direction or size of daily movement is different than that of one or both of its competitors, Public Stroage can evaluate what events or news impacting the company took place on that date. For example, if a new marketing campaign was started on a certain date, stock performance following that date can be evaluated to make some assumptions about the campaign's success.

Question 2: On which dates did Public Storage experience a change of at least 2.5% in their stock price (in either direction) and what happened to their competitors' stock prices on those dates?

Business Justification: Similar to the last query, this will allow Public Storage to evalute what possible outside factors impacted their stock in a significat manner (I assumed that a 2.5% change or greater was significant) and whether those factors similarly impacted their competitors.

SQL Features: CASE, GROUP BY, JOIN, CTE

```python
%%sql
WITH stock_comparison_cte AS (
	SELECT LEFT(psa_data.`Date`, 10) AS Date, 
		CASE
			WHEN SUM(psa_data.`Close`-psa_data.`Open`) = 0 THEN 'No Change'
			WHEN SUM(psa_data.`Close`-psa_data.`Open`) > 0 THEN 'Up'
			ELSE 'Down'
		END AS PSA_Daily_Movement,
		CONCAT(ROUND(SUM(((psa_data.`Close`-psa_data.`Open`)/psa_data.`Open`*100)), 2), '%') AS PSA_Daily_Change, 
		CASE
			WHEN SUM(cube_data.`Close`-cube_data.`Open`) = 0 THEN 'No Change'
			WHEN SUM(cube_data.`Close`- cube_data.`Open`) > 0 THEN 'Up'
			ELSE 'Down'
		END AS CUBE_Daily_Movement,
		CONCAT(ROUND(SUM(((cube_data.`Close`-cube_data.`Open`)/cube_data.`Open`*100)), 2), '%') AS CUBE_Daily_Change, 
		CASE
			WHEN SUM(lsi_data.`Close`-lsi_data.`Open`) = 0 THEN 'No Change'
			WHEN SUM(lsi_data.`Close`- lsi_data.`Open`) > 0 THEN 'Up'
			ELSE 'Down'
		END AS LSI_Daily_Movement,
		CONCAT(ROUND(SUM(((lsi_data.`Close`-lsi_data.`Open`)/lsi_data.`Open`*100)), 2), '%') AS LSI_Daily_Change
	FROM psa_data
	JOIN cube_data
		ON psa_data.`Date` = cube_data.`Date`
	JOIN lsi_data
		ON psa_data.`Date` = lsi_data.`Date`
	GROUP BY psa_data.`Date`
)
SELECT *
FROM stock_comparison_cte
HAVING PSA_Daily_Change >= 2.5 OR PSA_Daily_Change <= -2.5;
```

Question 2 Recommendations: While this query's results are similar to that of the first query, this result set is significanlty shorter and, therefore, is easier to read and work with. These data points should be what is used for impact anaylsis since there are much fewer dates to check here. Furthermore, if Public Storage decides on a value other than 2.5% as the threshold at which daily changes are considered significant, it can be easily updated here and the results will reflect the change.

Question 3: How has Public Storage's stock faired in comparison to its competitors since the first reported case of COVID-19 in the United States on January 19th, 2020? Compare the average of all daily changes starting on that date.

Business Justification: This query builds on the previous two queries to allow Public Storage to assess their performance against their competitors since the COVID-19 pandemic reached the United States.

SQL Features: CASE, GROUP BY, JOIN, CTE, VIEW, Window Function

```python
%%sql
CREATE OR REPLACE VIEW stock_comparison_significant_daily_movement AS 
	WITH stock_comparison_cte AS (
		SELECT LEFT(psa_data.`Date`, 10) AS Date, 
			CASE
				WHEN SUM(psa_data.`Close`-psa_data.`Open`) = 0 THEN 'No Change'
				WHEN SUM(psa_data.`Close`-psa_data.`Open`) > 0 THEN 'Up'
				ELSE 'Down'
			END AS PSA_Daily_Movement,
			CONCAT(ROUND(SUM(((psa_data.`Close`-psa_data.`Open`)/psa_data.`Open`*100)), 2), '%') AS PSA_Daily_Change, 
			CASE
				WHEN SUM(cube_data.`Close`-cube_data.`Open`) = 0 THEN 'No Change'
				WHEN SUM(cube_data.`Close`- cube_data.`Open`) > 0 THEN 'Up'
				ELSE 'Down'
			END AS CUBE_Daily_Movement,
			CONCAT(ROUND(SUM(((cube_data.`Close`-cube_data.`Open`)/cube_data.`Open`*100)), 2), '%') AS CUBE_Daily_Change, 
			CASE
				WHEN SUM(lsi_data.`Close`-lsi_data.`Open`) = 0 THEN 'No Change'
				WHEN SUM(lsi_data.`Close`- lsi_data.`Open`) > 0 THEN 'Up'
				ELSE 'Down'
			END AS LSI_Daily_Movement,
			CONCAT(ROUND(SUM(((lsi_data.`Close`-lsi_data.`Open`)/lsi_data.`Open`*100)), 2), '%') AS LSI_Daily_Change
		FROM psa_data
		JOIN cube_data
			ON psa_data.`Date` = cube_data.`Date`
		JOIN lsi_data
			ON psa_data.`Date` = lsi_data.`Date`
		GROUP BY psa_data.`Date`
	)
	SELECT *
	FROM stock_comparison_cte
	WHERE Date >= '2020-01-19';

SELECT
    CONCAT(ROUND(AVG(PSA_Daily_Change) OVER (), 2), '%') AS PSA_AVG_Daily_Change,
    CONCAT(ROUND(AVG(CUBE_Daily_Change) OVER (), 2), '%') AS CUBE_AVG_Daily_Change,
    CONCAT(ROUND(AVG(LSI_Daily_Change) OVER (), 2), '%') AS LSI_AVG_Daily_Change
FROM stock_comparison_significant_daily_movement
LIMIT 1;
```

Question 3 Recommendations: This final query further extends the work from the previous two queries. By narrowing down the results to the average daily change in stock price for each company since January 19th, 2020, the result set is much more actionable. Clearly, each company's stock has remained relatively unchanged since the COVID-19 pandemic started in the United States. With that said, however, Public Storage has clearly faired better than the competition since it is the only company to have gained value over this time period. This means that the company should stay the course in how they are responding to the pandemic and, if any special inititives were enacted to specifically address pandemic related challenges, they seem to be working.
